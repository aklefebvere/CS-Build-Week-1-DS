```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("iris.csv", header=None)
```

```python
df.head()
```

```python
x = np.array(df.drop(columns=[4]))
y = np.array(df[4])

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
X_train_df = pd.DataFrame(X_train)
X_train_df[4] = y_train
```

```python
X_train_df
```

```python
setosa = X_train_df[X_train_df[4] == 'Iris-setosa'][[0,1]]

versicolor = X_train_df[X_train_df[4] == 'Iris-versicolor'][[0,1]]

virginica = X_train_df[X_train_df[4] == 'Iris-virginica'][[0,1]]
```

```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 150
# plt.figure(figsize=(15,15))
ax = plt.gca()
circle = plt.Circle((X_test[1][0], X_test[1][1]), 0.55, color="Red", fill=False)
plt.scatter(setosa[0],setosa[1], label='Iris-setosa')
plt.scatter(versicolor[0], versicolor[1], label='Iris-versicolor')
plt.scatter(virginica[0], virginica[1], label='Iris-virginica')
plt.scatter(X_test[1][0], X_test[1][1], c='Black', label='Test data')
ax.add_artist(circle)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
ax.set_ylim([2.0, 4.55])
plt.title("K-Nearest Neighbor with k = 5")
plt.legend();
```
